```python
#Declaring a list
my_list = ['jordan', 'ellesse','nike','adidas','Vans off the wall','Off white' 'Cactus Jack', 'Palm angles','Dior','Essentials fear of god']
print(my_list)

#Accessing the list elements with indexing
print(my_list[1])

#Accessing the list elements with negative indexing
print(my_list[-1])

#looping through the list elements
for x in my_list:
    print(x)

#Checking or the list length
print(len(my_list))

#Declaring a set
my_set = {'jordan', 'ellesse','nike','adidas','Vans off the wall','Off white' 'Cactus Jack', 'Palm angles','Dior','Essentials fear of god'}
print(my_set)

#Accessing the set elements with indexing
#print(my_set[1])

#Accessing the set elements with negative indexing
#print(my_set[-1])

#looping through the set elements
for y in my_set:
    print(y)

#Checking the set length
print(len(my_set))

#Checking the data type
print(type(my_set))

#Declaring a python3 tuple
my_tuple = ('jordan', 'ellesse','nike','adidas','Vans off the wall','Off white' 'Cactus Jack', 'Palm angles','Dior','Essentials fear of god')
print(my_tuple)

#Accessing the tuple elements with indexing
print(my_tuple[1])

#Accessing the tuple elements with negative indexing
print(my_tuple[-1])

#looping through the tuple elements
for z in my_tuple:
    print(z)

#Checking or the tuple length
print(len(my_tuple))


#Declaring a Dictionary array
my_dict = {
    "brand":"Apple",
    "model":"ipad air",
    "year":2022
}

#Outing the Dictionary
print(my_dict)

#Looping through the Dictionary array
for x in my_dict:
    print(x)

#Accessing the Dictionary array
print(my_dict["brand"])

Kabwe_family = {
  "Mother" : {
    "name" : "Catherine Chiluba",
    "year" : 1978
  },
  "firstborn" : {
    "name" : "Mfuta Kabwe",
    "year" : 2000
  },
  "lastborn" : {
    "name" : "Chishibe Kabwe",
    "year" : 2004
  }
}
print(Kabwe_family)
```
